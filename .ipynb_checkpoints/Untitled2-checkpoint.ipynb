## 2-1 Load digiets 손글씨 분류기

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# (2) 데이터 준비, data/label지정하기
digits = load_digits()
digits_data = digits.data
digits_label = digits.target

# digit데이터의 feature 출력해보기
print(digits.feature_names)
# digit데이터의 타겟 이름 출력해보기
print(digits.target_names)
# digit데이터의 키 이름 출력해보기
print(digits.keys())
# digit데이터의 describe 해보기
print(digits.DESCR+"\n")

# pandas로 출력해보기
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
print("pandas------------------------------------")
print(digits_df)
print("구분------------------------------------")
      
      
# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# (4) 모델 학습 및 예측
# Desicion Tree
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
print("Desicion Tree------------------------------------")
print(classification_report(y_test, y_pred))

#Random Forest 
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print("Random Forest------------------------------------")
print(classification_report(y_test, y_pred))

#SVM Model  support Vector Machine
from sklearn import svm
svm_model = svm.SVC()
print(svm_model._estimator_type)

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("SVM Model------------------------------------")
print(classification_report(y_test, y_pred))

#SGC Model Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()
print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print("SGC Model------------------------------------")
print(classification_report(y_test, y_pred))


#Logistic Regression
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print("Logistic Regression------------------------------------")
print(classification_report(y_test, y_pred))
```

## 2-1 Digits 결론
Descision Tree : 0.86
Random Forest : 0.96
SVM :0.99
SGC : 0.94
Logistoc Regression : 0.95

SVM이 가장 높고 Desicision Tree가 가장 낮게 나왔다.
Desicision Tree는 가장 단순한 모델로 성능이 낮게 나오리라 예상은 했지만 나머지 모델들에 대해서는 사실 예상하기가 쉽지가 않다. 

## 2-2 wine 

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

# (2) 데이터 준비, data/label지정하기
wine = load_wine()
wine_data = wine.data
wine_label = wine.target

# wine feature 출력해보기
print(wine.feature_names)
# wine 타겟 이름 출력해보기
print(wine.target_names)
# wine 키 이름 출력해보기
print(wine.keys())
# wine describe 해보기
print(wine.DESCR+"\n")

# pandas로 출력해보기
digits_df = pd.DataFrame(data=wine_data, columns=wine.feature_names)
print("pandas------------------------------------")
print(digits_df)
print("구분------------------------------------")
      

# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# (4) 모델 학습 및 예측
# Desicion Tree
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
print("Desicion Tree------------------------------------")
print(classification_report(y_test, y_pred))

#Random Forest 
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print("Random Forest------------------------------------")
print(classification_report(y_test, y_pred))

#SVM Model  support Vector Machine
from sklearn import svm
svm_model = svm.SVC()
print(svm_model._estimator_type)

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("SVM Model------------------------------------")
print(classification_report(y_test, y_pred))

#SGC Model Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()
print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print("SGC Model------------------------------------")
print(classification_report(y_test, y_pred))

#Logistic Regression
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print("Logistic Regression------------------------------------")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

# (2) 데이터 준비, data/label지정하기
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target

# breast_cancer feature 출력해보기
print(breast_cancer.feature_names)
# breast_cancer 타겟 이름 출력해보기
print(breast_cancer.target_names)
# breast_cancer 키 이름 출력해보기
print(breast_cancer.keys())
# breast_cancer describe 해보기
print(breast_cancer.DESCR)

# pandas로 출력해보기
digits_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names)
print("pandas------------------------------------")
print(digits_df)
print("구분------------------------------------")
      

# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# (4) 모델 학습 및 예측
# Desicion Tree
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
print("Desicion Tree------------------------------------")
print(classification_report(y_test, y_pred))

#Random Forest 
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print("Random Forest------------------------------------")
print(classification_report(y_test, y_pred))

#SVM Model  support Vector Machine
from sklearn import svm
svm_model = svm.SVC()
print("SVM Model------------------------------------")
print(svm_model._estimator_type)

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("SVM Model------------------------------------")
print(classification_report(y_test, y_pred))

#SGC Model Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()
print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print("SGC Model------------------------------------")
print(classification_report(y_test, y_pred))

#Logistic Regression
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print("Logistic Regression------------------------------------")
print(classification_report(y_test, y_pred))
```
