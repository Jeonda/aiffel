 - 지도상 월별 유동인구와 코로나확진자 수를 GIF파일로 변환하여 확인해보자

### Prerequisites

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font', family='NanumBarunGothic')  

f_subway = pd.read_csv('aiffel/covid/subway_lines.csv') #역사 정보
f_subway.head()
f_subway_info=f_subway[['역이름', '위도', '경도', '호선']]
f_subway_info.sort_values('호선')
```

```python
import folium

map_osm = folium.Map(location=[37.555052,126.969138], zoom_start=12) #지도 초기 로딩위치

for item in f_subway_info.index :
  lat = f_subway_info.loc[item, '위도']
  long = f_subway_info.loc[item, '경도']
  folium.CircleMarker([lat, long], radius=1, popup=f_subway_info.loc[item,'역이름'], color = 'blue', fill = True).add_to(map_osm)

map_osm
```
